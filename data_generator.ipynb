```python
import requests
import pandas as pd
```

```python
def data_gen_from_api(symbol, dates, api = "ABCD1234EFGH5678"):
    df_dict = {}
    for _ in dates:
        url = f"https://www.alphavantage.co/query?function=TIME_SERIES_INTRADAY&symbol={symbol}&interval=5min&month={_}&outputsize=full&apikey={api}"
        r = requests.get(url)
        data = r.json()
    
        df = pd.DataFrame(columns = ["date", "open", "high", "low", "close", "volume"])
        df["date"] = data["Time Series (5min)"].keys()
        df["open"] = [data["Time Series (5min)"][i]["1. open"] for i in df["date"]]
        df["high"] = [data["Time Series (5min)"][i]["2. high"] for i in df["date"]]
        df["low"] = [data["Time Series (5min)"][i]["3. low"] for i in df["date"]]
        df["close"] = [data["Time Series (5min)"][i]["4. close"] for i in df["date"]]
        df["volume"] = [data["Time Series (5min)"][i]["5. volume"] for i in df["date"]]
        df_dict[_] = df.sort_values(by = "date").reset_index(drop = True)

    return pd.concat(tuple(df_dict.values()))
```
